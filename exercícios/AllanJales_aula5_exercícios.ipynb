````python
#Exercício 6

import math
import ColabTurtle.Turtle
joana = ColabTurtle.Turtle
joana.initializeTurtle()
joana.speed(7)

def draw_star():
    for i in range(5):
        joana.left(144)
        joana.forward(200)

#Estrela esquerda superior
joana.penup()
joana.left(45)
joana.forward(250)
joana.left(45)
joana.forward(200)
joana.pendown()
draw_star()

#Estrela direita superior
joana.penup()
joana.backward(450)
joana.pendown()
draw_star()

#Estrela direita inferior
joana.penup()
joana.left(90)
joana.forward(250)
joana.right(90)
joana.pendown()
draw_star()

#Estrela esquerda inferior
joana.penup()
joana.forward(450)
joana.pendown()
draw_star()
````

````python
#Exercício 1

import math
import ColabTurtle.Turtle
joana = ColabTurtle.Turtle
joana.initializeTurtle()
joana.speed(10)

def polygon(turtle, length, n, percent):
    alfa = 360*percent/n
    for i in range(n):
        turtle.forward(length)
        turtle.left(alfa)

def circle(turtle, r, percent = 1):
    circumference = 2*math.pi*r
    n = 10
    length = int(circumference/n)
    polygon(turtle, length, n, percent)

def petal(percent):
    r = 10
    angle = 180 - percent*360

    circle(joana, r, percent)
    joana.left(angle)
    circle(joana, r, percent)

#Flor da esquerda (7 pétalas)
joana.penup()
joana.left(90)
joana.forward(200)
joana.pendown()
for i in range(7):
    petal(1/5)
    joana.left(56.5)

#Flor do meio (10 pétalas)
joana.penup()
joana.left(180)
joana.forward(200)
joana.pendown()
for i in range(10):
    petal(1/5)
    joana.left(360/5)

#Flor da direita (20 pétalas)
joana.penup()
joana.forward(200)
joana.pendown()
for i in range(20):
    petal(1/10)
    joana.left(126)
````

````python
#Exercício 2

import math
import ColabTurtle.Turtle
joana = ColabTurtle.Turtle
joana.initializeTurtle()
joana.speed(10)

def triangle_isosceles(turtle, angle, height):
    turn_angle = 180 - (180 - angle)/2
    base_size = int(math.tan(math.radians(angle/2))*height)*2
    cathetus_size = int(height/math.cos(math.radians(angle/2)))

    turtle.forward(cathetus_size)
    turtle.left(turn_angle)
    turtle.forward(base_size)
    turtle.left(turn_angle)
    turtle.forward(cathetus_size)
    turtle.left(180)

def n_tagono(n, height):
    for i in range(n):
        triangle_isosceles(joana, 360/n, height)

#Esquerda
joana.penup()
joana.left(90)
joana.forward(200)
joana.pendown()
n_tagono(5, 75)

#Meio
joana.penup()
joana.left(180)
joana.forward(200)
joana.left(90)
joana.pendown()
n_tagono(6, 75)

#Direita
joana.penup()
joana.right(90)
joana.forward(200)
joana.pendown()
n_tagono(7, 75)
````
